```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/majestic-lion-roaring-stockcake.jpg")

if image is None:
    print("Issue with image loading.")
else:
    print("Image loaded successfully.")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_greyscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
hist, bins = np.histogram(image_greyscale.flatten(), 256, [0,256])

# Perform histogram equalization

equialised_image = cv2.equalizeHist(image_greyscale)

equalised_hist, bins = np.histogram(equialised_image.flatten(), 256, [0,256])

# plot the results
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(image_greyscale, cmap='grey')
plt.title('Original image')

plt.subplot(2,2,2)
plt.plot(hist, color='blue')
plt.title('Original image histogram')

plt.subplot(2,2,3)
plt.imshow(equialised_image, cmap='grey')
plt.title('Original image')

plt.subplot(2,2,4)
plt.plot(equalised_hist, color='green',)
plt.title('Equalised image histogram')

plt.tight_layout()
plt.show()
```

```python
image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

image_yuv[:,:,0] = cv2.equalizeHist(image_yuv[:, :, 0])
#  y channel handles the brightness so we equalize it
equalised_color_image = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR)

equalised_color_image_RGB = cv2.cvtColor(equalised_color_image, cv2.COLOR_BGR2RGB)

# plot the results
plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('Original image')

plt.subplot(1,2,2)
plt.imshow(equalised_color_image_RGB)
plt.title('Equalised rgb image')

plt.tight_layout()
plt.show()
```
