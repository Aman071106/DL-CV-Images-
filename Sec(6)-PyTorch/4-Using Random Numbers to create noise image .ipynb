```python
import torch
```

## generate random integers

```python
# torch.randint(low, high, size)

random_integers = torch.randint(-10,10,(5,5))
print(random_integers)
```

```python
# generate random numbers froma normal distribution

# torch.normal(mean, std, size)

random_numbers_normal = torch.normal(0.5,2, (5,))
print(random_numbers_normal)
```

## Creating noise image using pytorch

```python
import matplotlib.pyplot as plt

# h,w of image 256x256

uniform_noise = torch.rand((256,256))

plt.imshow(uniform_noise.numpy(), cmap='grey')
plt.title("Uniform Noise Image")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

# h,w,c of image 256x256X3

uniform_noise_rgb = torch.rand((256,256,3))*255

plt.imshow(uniform_noise_rgb.numpy(),)
plt.title("Uniform Noise Image RGB")
plt.axis('off')
plt.show()
```

```python
print("Shape", uniform_noise_rgb.shape)
print("Data type", uniform_noise_rgb.dtype)
print("Min value", torch.min(uniform_noise_rgb))
print("Max value", torch.max(uniform_noise_rgb))
```
