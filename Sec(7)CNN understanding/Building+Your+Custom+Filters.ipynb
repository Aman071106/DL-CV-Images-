# Building Custom Filters

```python
# Functions for Downloading & Displaying Images
# Import necessary packages
import numpy as np
import cv2
from urllib.request import urlopen
from matplotlib import pyplot as plt

def download_image(url):
    """
    Downloads an image from the given URL and returns it as an OpenCV image object.

    Parameters:
        url (str): The URL of the image to download.

    Returns:
        np.ndarray: Image in OpenCV format (BGR).
    """
    # Download the image and convert it to a NumPy array
    with urlopen(url) as resp:
        image_data = np.asarray(bytearray(resp.read()), dtype="uint8")
    # Decode the image data to OpenCV format
    image = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
    return image

def display_image(image, cmap=None):
    """
    Displays an OpenCV image using matplotlib.

    Parameters:
        image (np.ndarray): The OpenCV image object to display.
        cmap (str, optional): Colormap for displaying grayscale images. Defaults to None.
    """
    # Convert the image from BGR to RGB for correct color display in matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb, cmap=cmap)
    plt.axis('off')  # Hide axis for a cleaner look
    plt.show()

# # Example usage
# url = 'https://example.com/image.jpg'  # Replace with the actual URL
# image = download_image(url)
# display_image(image)
```

```python
url = 'https://i.etsystatic.com/20636120/r/il/7724a4/2806797038/il_570xN.2806797038_7eln.jpg'  # Image URL
image = download_image(url)
display_image(image)
```

## Vertical Edge Detector

```python
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]])

edge_detector = cv2.filter2D(image,-1,kernel)
cv2.imshow("",edge_detector)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
