[Explore YOLO Models](https://docs.ultralytics.com/tasks/detect/#models)

```python
import cv2
from matplotlib import pyplot as plt

def show_image(image_path):
    """
    Displays an image using matplotlib (works in Google Colab/Jupyter).
    """
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found!")
        return

    # Convert BGR (OpenCV) to RGB (matplotlib)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axes
    plt.title('Predicted Image')
    plt.show()


```

Nano Model

```python
# Run inference on an image with YOLO11n
!yolo predict model=yolo11n.pt source='https://images.pexels.com/photos/169198/pexels-photo-169198.jpeg'
```

```python
show_image('runs/detect/predict/pexels-photo-169198.jpg')
```
